```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 10  #радиус гиперсферы
volumeRatio = [] #массив для хранения отношения объемов для измерений от 2 до 10
for i in range(2, 10):  #измерения
    allPoints = np.random.random(size=(10000, i)) * r
    pointsInsideTheSphere = 0 #количество точек гиперкуба, попавших в гиперсферу
    for point in allPoints:
        formulaSphere = 0
        for j in point:
            formulaSphere += j ** 2 #сумма квадратов измерений   
        if formulaSphere <= r**2:
            pointsInsideTheSphere += 1
    volumeRatio.append(pointsInsideTheSphere / 10000)
print(volumeRatio)
plt.plot(range(2, 10), volumeRatio)
plt.show()
```
