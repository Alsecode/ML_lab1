```python
#Задание 1.3
#Для датасета California Housing:
#Разбейте датасет на тренировочную, валидационную и тестовую выборку
#Проведите преобразование категориального признака ocean_proximity через OneHot или Dummy-кодировку
#Замените признаки total_rooms и total_bedrooms на average_rooms и average_bedrooms (поделив на households).
#В признаке average_bedrooms (total_bedrooms) есть отсутствующие значения. 
#Определите число экземпляров данных, для которых этот признак отсутствует. 
#Придумайте и обоснуйте стратегию заполнения пропусков в этой задаче. Заполните пропуски.
#Нормализуйте признаки longitude и latitude (сделайте так, чтобы каждый признак имел среднее значение 0 и дисперсию 1 внутри обучающей выборки)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
housing_df = pd.read_csv('CSV_files/housing.csv')
housing_df
```

```python
#Преобразование категориального признака ocean_proximity через Dummy-кодировку
DUMMY_SEP = '__'  #Сепаратор

housing_df = pd.get_dummies(housing_df, columns = ['ocean_proximity'], prefix_sep = DUMMY_SEP)
housing_df
```

```python
#Замена признаков total_rooms и total_bedrooms на average_rooms и average_bedrooms (поделив на households)
housing_df['average_rooms'] = housing_df['total_rooms'] / housing_df['households']
housing_df['average_bedrooms'] = housing_df['total_bedrooms'] / housing_df['households']

housing_df = housing_df.drop('total_rooms', axis = 1)
housing_df = housing_df.drop('total_bedrooms', axis = 1)

housing_df.head(294)
#Пример: признак отсутствует у поля 290
```

```python
#Определить число экземпляров данных, для которых отсутствует признак average_bedrooms
average_bedrooms_blank = sum(housing_df['average_bedrooms'].isnull())
average_bedrooms_blank
```

```python
#Придумать и обосновать стратегию заполнения пропусков. Заполнить пропуски.

#Новый дата-сет без данных с отсутствующим 'average_bedrooms', а также без признака 'median_house_value'
housing_df_new = housing_df.dropna(axis = 0)  
housing_df_new = housing_df_new.drop('median_house_value', axis = 1)

housing_df_new
```

```python
#Разбиение нового дата-сета на тренировочную и тестовую выборку
housing_train_df_new, housing_test_df_new =  train_test_split(housing_df_new, test_size = 0.1)

x_train_df = housing_train_df_new.drop('average_bedrooms', axis = 1)  #Вход для тренировочной
y_train_df = housing_train_df_new['average_bedrooms']   #Выход для тренировочной

x_test_df = housing_test_df_new.drop('average_bedrooms', axis = 1)   #Вход для тестовой
y_test_df = housing_test_df_new['average_bedrooms']    #Выход для тестовой

#Модель линейной регрессии
linear_regression_model = LinearRegression()

#Обучение модели с помощью fit
linear_regression_model.fit(x_train_df, y_train_df)
```

```python
#Абсолютная оценка регрессионной модели
mean_absolute_error(y_train_df, linear_regression_model.predict(x_train_df)) 
```

```python
#Среднеквадратическая оценка регрессионной модели 
mean_squared_error(y_train_df, linear_regression_model.predict(x_train_df)) 
```

```python
#Записи, для которых отсутсвтует признак 'average_bedrooms' (только столбец 'average_bedrooms', булевые значения)
housing_df_blank_bool = housing_df['average_bedrooms'].isnull()
housing_df_blank_bool
```

```python
#Полные записи, для которых отсутсвтует признак 'average_bedrooms'
housing_df_blank = housing_df.loc[housing_df_blank_bool]
housing_df_blank
```

```python
#Отбрасываем признаки 'median_house_value' и 'average_bedrooms'
housing_df_blank = housing_df_blank.drop(['median_house_value', 'average_bedrooms'], axis = 1)
housing_df_blank
```

```python
#Заполнение пропусков в задаче
#Массив полученных с помощью регрессии данных для заполнения 207 пустых значений
predictions_av_bedrooms = linear_regression_model.predict(housing_df_blank)
predictions_av_bedrooms
```

```python
housing_df.loc[housing_df_blank_bool, 'average_bedrooms'] = predictions_av_bedrooms
```

```python
#Разбиение датасета на тренировочную, валидационную и тестовую выборку:
housing_train_val, housing_test = train_test_split(housing_df, test_size = 0.1)
housing_train, housing_val = train_test_split(housing_train_val, test_size = 0.1)
```

```python
#Нормализация признаков longitude и latitude:
#(сделать так, чтобы каждый признак имел среднее значение 0 и дисперсию 1 внутри обучающей выборки)

#Для нормализации данных можно воспользоваться функцией StandardScaler из sklearn.preprocessing. 
#Данные преобразуются так, что их распределение имеет среднее значение 0 и стандартное отклонение 1 (корень из дисперсии).

housing_train[['longitude', 'latitude']] = StandardScaler().fit_transform(housing_train[['longitude', 'latitude']])

housing_test[['longitude', 'latitude']] = StandardScaler().fit_transform(housing_test[['longitude', 'latitude']])

housing_val[['longitude', 'latitude']] = StandardScaler().fit_transform(housing_val[['longitude', 'latitude']])

print("Для признака longitude. Среднее значение тренировочной выборки равно: " + str(housing_train['longitude'].to_numpy().mean(axis = 0)))
print("Для признака longitude. Стандартное отклонение тренировочной выборки равно: " + str(housing_train['longitude'].to_numpy().std(axis = 0)))
print("\n")
print("Для признака latitude. Среднее значение тренировочной выборки равно: " + str(housing_train['latitude'].to_numpy().mean(axis = 0)))
print("Для признака latitude. Стандартное отклонение тренировочной выборки равно: " + str(housing_train['latitude'].to_numpy().std(axis = 0)))
```
