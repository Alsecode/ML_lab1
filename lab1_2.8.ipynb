```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def transformation(matrix):
    new_matrix = matrix.copy()
    for i in range(matrix.shape[1]):
        d = matrix[:, i]  #столбцы матрицы
        averageValue = np.mean(d)  #среднее значение в столбце
        deviation = np.std(d)  #стандартное отклонение в столбце
        new_matrix[:,i] = (d - averageValue) / deviation  #заполнение новой матрицы
    return new_matrix
```

```python
# отрисовка результата преобразования точек для d = 2
points = np.random.multivariate_normal(mean=[1,2],cov=[[2,1],[1,3]], size = 1000)
new_points = transformation(points)
fig, ax = plt.subplots()
ax.scatter(points[:, 0], points[:, 1], color = 'red', label = 'Оригинальные точки для d=2')
ax.scatter(new_points[:, 0], new_points[:, 1], color = 'blue', label = 'Результат преобразования для d=2')
plt.legend()
plt.show()
```
