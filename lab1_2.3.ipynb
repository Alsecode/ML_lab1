```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pic(r):
    figure = plt.figure()
    ax = figure.add_subplot(1,1,1)
    r = float(r)
    ax.set_xlim(0, 2 * r + r * 0.25)  #Масштаб по x
    ax.set_ylim(0, 2 * r + r * 0.25)  #Масштаб по y
    circle = plt.Circle((r + r*0.125, r + r*0.125), radius=r, edgecolor='green', facecolor='None')  #Масштабирование центра окружности и рисование окружности радиуса r
    ax.add_patch(circle)
    plt.show()
```

```python
r = input("Введите радиус: ") 
pic(r)
```
