```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def transformation_plot(points, A):
    new_points = points.copy()
    for i in range(len(points)):
        new_points[i] = np.dot(A, points[i])
    return new_points
```

```python
points = np.array([[7.7, 2], [2, 3], [10, -5], [10.8, 7], [-3.3, 5.8]])
```

```python
A = np.array([[10, 12], [13, 19]])
```

```python
new_points = transformation_plot(points, A)
```

```python
for i in range (len(points)):
    plt.plot(points[i][0], points[i][1])
    plt.scatter(points[i][0], points[i][1], color='yellow', s=70, marker='o')
    plt.text(points[i][0], points[i][1], i + 1, color='red')
    plt.title("Точки до преобразования")
plt.show()

for i in range (len(new_points)):
    plt.plot(new_points[i][0], new_points[i][1])
    plt.scatter(new_points[i][0], new_points[i][1], color='yellow', s=70, marker='o')
    plt.text(new_points[i][0], new_points[i][1], i + 1, color='red')
    plt.title("Точки после преобразования")
plt.show()
```
