```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw():
    figure, ax = plt.subplots()
    x = np.linspace(-5, 5, 100)
    f = x - x**3 / 6 + x**5 / 120 - x**7 / 5040
    g = np.sin(x)
    ax.plot(x, f, label = "f(x)", color = "orange")
    ax.plot(x, g, label = "g(x)", color = "blue")
    plt.legend(loc = "upper center");
    plt.show()
```

```python
draw()
```
