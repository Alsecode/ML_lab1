```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def functions_graphs():
    x = np.linspace(-10,10,100)
    y = x**2 + 3 * x - 4
    dx = x[1] - x[0] #приращение x
    dy = np.gradient(y, dx) #приращение y
    
    plt.plot(x, y, color = 'yellow')
    plt.title("График оригинальной функции ")
    plt.show()
    
    plt.plot(x, dy, color = 'orange')
    plt.title("График производной (численный метод) ")
    plt.show()
    
    plt.plot(x, 2 * x + 3, color = 'red')
    plt.title("График производной (аналитический метод) ")
    plt.show()
```

```python
functions_graphs()
```
